## Exerscise 1 part (iv)

```python
a = 1
m = 1
kappa1 = 1
kappa2 = .8

def nn(k):
    return 2 * np.sqrt(1/m) * np.sqrt((kappa1 * np.sin(k*a/2) ** 2 + kappa2 * np.sin(k*a)**2))

k_vals = np.linspace(-np.pi/a, np.pi/a, 101)

fig, ax = plt.subplots()

ax.plot(k_vals, nn(k_vals));
ax.set_xlabel(r'$k$'); 
ax.set_ylabel('$\omega$');
ax.set_xlim(-2,2)
ax.set_title('Dispersion')
plt.xticks([-np.pi/a, 0, np.pi/a], [r'$-\pi/a$', 0, r'$\pi/a$']);
plt.tight_layout();

plt.show()
```
