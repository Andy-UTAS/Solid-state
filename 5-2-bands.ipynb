```python
def E(k_x, k_y):
    delta = np.array([-2*np.pi, 0, 2*np.pi])
    H = np.diag(
        ((k_x + delta)[:, np.newaxis]**2
        + (k_y + delta)[np.newaxis]**2).flatten()
    )
    return tuple(np.linalg.eigvalsh(H + 5)[:3])

E = np.vectorize(E, otypes=(float, float, float))

momenta = np.linspace(-2*np.pi, 2*np.pi, 100)
kx, ky = momenta[:, np.newaxis], momenta[np.newaxis, :]
bands = E(kx, ky)

# Extended Brillouin zone scheme
pad = .3
first_BZ = ((abs(kx) < np.pi + pad) & (abs(ky) < np.pi + pad))
second_BZ = (
    ((abs(kx) > np.pi - pad) | (abs(ky) > np.pi - pad))
    & ((abs(kx + ky) < 2*np.pi + pad) & (abs(kx - ky) < 2*np.pi + pad))
)
third_BZ = (
    (abs(kx + ky) > 2*np.pi - pad) | (abs(kx - ky) > 2*np.pi - pad)
)

bands[0][~first_BZ] = np.nan
bands[1][~second_BZ] = np.nan
#bands[2][~third_BZ] = np.nan

# Actually plotting

fig = go.Figure(
    data = [
        go.Surface(
            z=band / 5,
            # colorscale=color,
            opacity=opacity,
            showscale=False,
            hoverinfo='none',
            colorscale='Reds',
            x=momenta,
            y=momenta,
        )
        for band, opacity
        in zip(bands[:2],
               # ['#cf483d', '#3d88cf'],
               (1, 0.9))
    ],
    layout = go.Layout(
        title='Nearly free electrons in 2D',
        autosize=True,
        hovermode=False,
        margin=dict(
            t=50,
            l=20,
            r=20,
            b=50,
        ),
        scene=dict(
            yaxis={"title": "k_y"},
            xaxis={"title": "k_x"},
            zaxis={"title": "E"},
        )
    )
)

# html file
if True:
    py.plot(fig, filename='5-2-nearly-free.html')

py.iplot(fig, show_link=False)
```
